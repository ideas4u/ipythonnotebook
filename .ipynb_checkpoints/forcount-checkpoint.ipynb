```python
project_data.columns
colName = ['序号', '项目名称', '项目编码', '需求单位', '建设单位', '建设部门', '一级专业', '立项编制人', '项目经理',
       '管理属性', '工程进度', '立项提交时间', '实际立项批复时间', '计划立项批复时间', '立项批复总投资',
       '立项批复本年CAPEX', '可研提交时间', '实际可研批复时间', '计划可研批复时间', '实际提交采购时间', '计划提交采购时间',
       '实际合同小签时间', '计划合同小签时间', '实际合同大签时间', '计划合同大签时间', '实际到货时间', '计划到货时间',
       '实际开工时间', '计划开工时间', '设计提交时间', '实际设计批复时间', '计划设计批复时间', '设计批复总投资',
       '设计批复本年CAPEX', '实际初验时间', '计划初验时间', '实际资产暂估时间', '计划资产暂估时间', '实际终验时间',
       '计划终验时间', '实际决算审计完成时间', '计划决算审计完成时间', '实际决算上报时间', '计划决算上报时间',
       '实际决算批复及固定资产调整时间', '计划决算批复及固定资产调整时间', '结算送审率','pd结算送审率']
project_data_new = project_data.reindex(columns=colName,method='ffill')
```

```python
project_data_new.head()
jieshu_data.head()
```

```python
result = pd.merge(project_data_new,jieshu_data,how='left',on ='项目编码')
result
#project_data_new
```

```python
grouped_base = result.groupby('建设单位')
grouped_base.count()
```

```python
selet_result = result[result['结算送审数/需结算送审数%']==1]
selet_result
```

```python
grouped = selet_result.groupby('建设单位')
grouped.count()
```

```python
finalResult = pd.merge(grouped_base.count(),grouped.count(),how='left',left_index=True, right_index=True)
finalResult = finalResult[['需求单位_x','结算送审数/需结算送审数%_y']]
finalResult = finalResult.fillna(0)
finalResult.columns
finalResultName = ['需求单位_x', '结算送审数/需结算送审数%_y','送审及时率']
finalResult = finalResult.reindex(columns=finalResultName)
finalResult['送审及时率'] = finalResult['结算送审数/需结算送审数%_y']/finalResult['需求单位_x']
#finalResult.to_excel('outFileResult.xlsx')
finalResult['送审及时率'] = finalResult['送审及时率']*100/100
finalResult
```

```python
result.to_excel('outFile1.xlsx',sheet_name='Sheet1')
```
