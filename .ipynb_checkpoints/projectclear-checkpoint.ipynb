```python
os.getcwd()
project_data_list=pd.read_excel('宽带-在建工程清理全省项目明细数据20151109.xlsx',sheetname='分公司清理项目明细样表',header=2)
project_data_list.index= project_data_list['项目编号']
project_data_list.head()project_data_list.head()
```

```python
project_data_list['项目进度最新状态.1'].drop_duplicates()
```

```python
sheji_data_list = pd.read_excel('导出的审计信息.xlsx',sheetname='导出的审计信息')
```

```python
sheji_data_list.rename(columns={'项目编码':'项目编号'},inplace=True )
sheji_data_list.index = sheji_data_list['项目编号']
sheji_data_list
```

```python
project_data_list_merged=pd.merge(project_data_list,sheji_data_list,how='left',left_on='项目编号',right_on='项目编号')
#合并二个表格，相当于VLOOKUP.
```

```python
#对完成送审的项目进行统计，
project_data_list_merged_completed=project_data_list_merged[project_data_list_merged['结算送审数/需结算送审数%'] == 1]
```

```python
project_data_list_merged.columns
```

```python
#对不用送审的项目进行也统计
project_data_list_merged_closed=project_data_list_merged[project_data_list_merged['项目进度最新状态.1'] == '结项（即项目关闭）'][project_data_list_merged['结算送审数/需结算送审数%'] <1]
```

```python
project_data_list_merged_closed
```

```python
#对完成送审项目与结项项目进行合并
project_data_list_merged_both_completed=pd.concat([project_data_list_merged_completed,project_data_list_merged_closed],join='outer')
```

```python
project_data_list_merged_both_completed.groupby('分公司').count()
```

```python
#完成送审数统计
project_data_list_songsheng_completed=project_data_list_merged_both_completed.groupby('分公司').count()[['项目编号','结算送审数/需结算送审数%']]
```
